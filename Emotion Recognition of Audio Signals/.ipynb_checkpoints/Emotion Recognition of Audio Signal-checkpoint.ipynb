```python
import librosa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
from PIL import Image
import pathlib
import csv
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
import keras
from keras import layers
from keras.models import Sequential
import warnings
import scipy

from numpy.linalg import inv
warnings.filterwarnings('ignore')

def feature_extraction_windowed(savePath, emotions = 'arrabbiato calmo cupo drammatico felice funky motivante romantico triste vivace'):

  header = ""

  for i in range(0,12):
      header += 'chroma_mean_'+str(i)+' '
  for i in range(0,12):
      header += 'chroma_std_'+str(i)+' '

  for i in range(0,13):
      header += 'mfccs_mean_'+str(i)+' '
  for i in range(0,13):
      header += 'mfccs_std_'+str(i)+' '

  header += 'cent_mean cent_std cent_skew '

  for i in range(0,7):
      header += 'contrast_mean_'+str(i)+' '
  for i in range(0,7):
      header += 'contrast_std_'+str(i)+' '

  header += 'rolloff_mean rolloff_std rolloff_skew zrate_mean zrate_std zrate_skew tempo emotion'

  header = header.split()
  file = open(savePath, 'w', newline='')
  with file:
      writer = csv.writer(file)
      writer.writerow(header)
  emotions = emotions.split()
  count = 0
  for g in emotions:
      for filename in os.listdir(f'Tracks/{g}'):
          songname = f'Tracks/{g}/{filename}'
          songname = os.fspath(songname)
          track, sr = librosa.load(songname, mono=True)
          
          #normalized_track = librosa.util.normalize(track)
          count += 1
          print(str(count) + ")Nuova traccia: " + songname)
          #print("Dimensione traccia: " + str(track.shape))
          frame_len, hop_len = 337500, 225000 #15s window con 10s hop
          frames = librosa.util.frame(track, frame_length=frame_len, hop_length=hop_len, axis = 0)
        
          #print(frames.shape)
            
          for y in frames:
            
            y_harmonic, y_percussive = librosa.effects.hpss(y)

            tempo, beat_frames = librosa.beat.beat_track(y=y_harmonic, sr=sr)
            tempo = int(tempo)
            rmse = librosa.feature.rms(y=y)
            #rmse = librosa.util.normalize(rmse)

            chroma = librosa.feature.chroma_cens(y=y_harmonic, sr=sr)
            #chroma = librosa.util.normalize(chroma)

            cent = librosa.feature.spectral_centroid(y=y, sr=sr)
            #cent = librosa.util.normalize(cent)
            spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
            #spec_bw = librosa.util.normalize(spec_bw)

            contrast = librosa.feature.spectral_contrast(y=y_harmonic,sr=sr)
            #contrast = librosa.util.normalize(contrast)

            rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
            #rolloff = librosa.util.normalize(rolloff)
            zcr = librosa.feature.zero_crossing_rate(y_harmonic)
           # zcr = librosa.util.normalize(zcr)

            mfccs = librosa.feature.mfcc(y=y_harmonic, sr=sr, n_mfcc=13)
           # mfccs = librosa.util.normalize(mfccs)

            chroma_mean=np.mean(chroma,axis=1)
            chroma_std=np.std(chroma,axis=1)

            #Generate the chroma Dataframe
            chroma_df=pd.DataFrame()
            for i in range(0,12):
                chroma_df['chroma_mean_'+str(i)]=chroma_mean[i]
            for i in range(0,12):
                chroma_df['chroma_std_'+str(i)]=chroma_mean[i]
            chroma_df.loc[0]=np.concatenate((chroma_mean,chroma_std),axis=0)


            mfccs_mean=np.mean(mfccs,axis=1)
            mfccs_std=np.std(mfccs,axis=1)
            #Generate the chroma Dataframe
            mfccs_df=pd.DataFrame()
            for i in range(0,13):
                mfccs_df['mfccs_mean_'+str(i)]=mfccs_mean[i]
            for i in range(0,13):
                mfccs_df['mfccs_std_'+str(i)]=mfccs_mean[i]
            mfccs_df.loc[0]=np.concatenate((mfccs_mean,mfccs_std),axis=0)

            cent_mean=np.mean(cent)
            cent_std=np.std(cent)
            cent_skew=scipy.stats.skew(cent,axis=1)[0]

            contrast_mean=np.mean(contrast,axis=1)
            contrast_std=np.std(contrast,axis=1)

            #Generate the chroma Dataframe
            contrast_df=pd.DataFrame()

            rolloff_mean=np.mean(rolloff)
            rolloff_std=np.std(rolloff)
            rolloff_skew=scipy.stats.skew(rolloff,axis=1)[0]

            spectral_df=pd.DataFrame()
            collist=['cent_mean','cent_std','cent_skew']
            for i in range(0,7):
                collist.append('contrast_mean_'+str(i))
            for i in range(0,7):
                collist.append('contrast_std_'+str(i))
            collist=collist+['rolloff_mean','rolloff_std','rolloff_skew']
            for c in collist:
                spectral_df[c]=0
            data=np.concatenate(([cent_mean,cent_std,cent_skew],contrast_mean,contrast_std,[rolloff_mean,rolloff_std,rolloff_skew]),axis=0)
            spectral_df.loc[0]=data

            zcr_mean=np.mean(zcr)
            zcr_std=np.std(zcr)
            zcr_skew=scipy.stats.skew(zcr,axis=1)[0]

            zrate_df=pd.DataFrame()
            zrate_df['zrate_mean']=0
            zrate_df['zrate_std']=0
            zrate_df['zrate_skew']=0
            zrate_df.loc[0]=[zcr_mean,zcr_std,zcr_skew]

            beat_df=pd.DataFrame()
            beat_df['tempo']=tempo
            beat_df.loc[0]=tempo

    
            to_append = pd.concat((chroma_df,mfccs_df,spectral_df,zrate_df,beat_df),axis=1).to_string(header=False,
                                                                                                      index=False,
                                                                                                      index_names=False)  
            #print(to_append)
            to_append += f' {g}'

            file = open(savePath, 'a', newline='')
            with file:
                writer = csv.writer(file)
                writer.writerow(to_append.split())
```

```python
feature_extraction_windowed('Dataset_triste_windowed(15s-10s).csv',"triste")
```

```python
import librosa
import pandas as pd
import matplotlib.pyplot as plt
import os
import pathlib
import numpy as np
import warnings

warnings.filterwarnings('ignore')

def features_analysis(songname):
    songname = os.fspath(songname)
    y, sr = librosa.load(songname, mono=True, duration=20)
    
    plt.figure()
    plt.subplot(3, 1, 1)
    plt.title('Waveform')
    librosa.display.waveshow(y, sr=sr)
    
    y_harmonic, y_percussive = librosa.effects.hpss(y)
    
    fig, (ax, ax2) = plt.subplots(nrows=2, sharex=True)
    ax.set(xlim=[6.0, 6.01], title='Sample view', ylim=[-0.5, 0.5])
    librosa.display.waveshow(y, sr=sr, ax=ax, marker='.', label='Full signal')
    librosa.display.waveshow(y_harmonic, sr=sr, alpha=0.5, ax=ax2, label='Harmonic')
    librosa.display.waveshow(y_percussive, sr=sr, color='r', alpha=0.5, ax=ax2, label='Percussive')
    ax.label_outer()
    ax.legend()
    ax2.legend()

    tempo, beat_frames = librosa.beat.beat_track(y=y_harmonic, sr=sr)
    tempo = int(tempo)
    
    S = librosa.magphase(librosa.stft(y, window=np.ones, center=False))[0]
    rms = librosa.feature.rms(y=y, S=S)
    #rmse = librosa.util.normalize(rmse)
    fig, ax = plt.subplots(nrows=2, sharex=True)
    times = librosa.times_like(rms)
    ax[0].semilogy(times, rms[0], label='RMS Energy')
    ax[0].set(xticks=[])
    ax[0].legend()
    ax[0].label_outer()
    librosa.display.specshow(librosa.amplitude_to_db(S, ref=np.max),
                             y_axis='log', x_axis='time', ax=ax[1])
    ax[1].set(title='log Power spectrogram')
    plt.show()

    chroma = librosa.feature.chroma_cens(y=y_harmonic, sr=sr)
    #chroma = librosa.util.normalize(chroma)

    cent = librosa.feature.spectral_centroid(y=y, sr=sr)
    #cent = librosa.util.normalize(cent)
    spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
    #spec_bw = librosa.util.normalize(spec_bw)

    contrast = librosa.feature.spectral_contrast(y=y_harmonic,sr=sr)
    #contrast = librosa.util.normalize(contrast)

    rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
    #rolloff = librosa.util.normalize(rolloff)
    zcr = librosa.feature.zero_crossing_rate(y_harmonic)
    # zcr = librosa.util.normalize(zcr)

    mfccs = librosa.feature.mfcc(y=y_harmonic, sr=sr, n_mfcc=13)
    # mfccs = librosa.util.normalize(mfccs)
```
