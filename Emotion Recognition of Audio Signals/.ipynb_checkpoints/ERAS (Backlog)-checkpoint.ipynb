# Implementazione algoritmo di classificazione

```python
def trainMls(dataset, indici = "ABCDEFGHIJ", neighbors = 5, test_size = 0.2):
  from sklearn.metrics import classification_report
  from sklearn.metrics import confusion_matrix
  from sklearn.metrics import accuracy_score
  from sklearn.neighbors import KNeighborsClassifier
  from sklearn.linear_model import LogisticRegression
  from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
  from sklearn.svm import SVC
  from sklearn import tree

  data = pd.read_csv(dataset)
  data.head()# Dropping unneccesary columns
  #data = data.drop(['filename'],axis=1)#Encoding the Labels
  emotion_list = data.iloc[:, -1]
  encoder = LabelEncoder()
  encoder = encoder.fit(indici)
  y = encoder.transform(emotion_list)#Scaling the Feature columns
  scaler = StandardScaler()
  #from sklearn.preprocessing import MinMaxScaler
  #scaler = MinMaxScaler()
  X = scaler.fit_transform(np.array(data.iloc[:, :-1], dtype = float))
  #Dividing data into training and Testing set
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= test_size)

  SVC_model = SVC()
  # KNN model requires you to specify n_neighbors,
  # the number of points the classifier will look at to determine what class a new point belongs to
  KNN_model = KNeighborsClassifier(n_neighbors=neighbors)

  SVC_model = SVC_model.fit(X_train, y_train)
  KNN_model = KNN_model.fit(X_train, y_train)

  SVC_prediction = SVC_model.predict(X_test)
  KNN_prediction = KNN_model.predict(X_test)

  import seaborn as sn
  import matplotlib.pyplot as plt

  print("SVC")
  print(classification_report(SVC_prediction, y_test, target_names = indici))
  df_cm = pd.DataFrame(confusion_matrix(SVC_prediction, y_test, normalize='true'), index = [i for i in indici],
                    columns = [i for i in indici])
  plt.figure(figsize = (10,7))
  sn.heatmap(df_cm, annot=True)
  plt.show()

  print("KNN")
  print(classification_report(KNN_prediction, y_test, target_names = indici))
  df_cm = pd.DataFrame(confusion_matrix(KNN_prediction, y_test, normalize='true'), index = [i for i in indici],
                    columns = [i for i in indici])
  plt.figure(figsize = (10,7))
  sn.heatmap(df_cm, annot=True)

  plt.show()
```

# Nuova versione di estrazione delle features con un totale di 74 features e l'analisi della traccia per intero (V2)

```python
import scipy

def feature_extraction(savePath, emotions = 'arrabbiato calmo cupo drammatico felice funky motivante romantico triste vivace'):

  header = ""

  for i in range(0,12):
      header += 'chroma_mean_'+str(i)+' '
  for i in range(0,12):
      header += 'chroma_std_'+str(i)+' '

  for i in range(0,13):
      header += 'mfccs_mean_'+str(i)+' '
  for i in range(0,13):
      header += 'mfccs_std_'+str(i)+' '

  header += 'cent_mean cent_std cent_skew '

  for i in range(0,7):
      header += 'contrast_mean_'+str(i)+' '
  for i in range(0,7):
      header += 'contrast_std_'+str(i)+' '

  header += 'rolloff_mean rolloff_std rolloff_skew zrate_mean zrate_std zrate_skew tempo emotion'

  header = header.split()
  file = open(savePath, 'w', newline='')
  with file:
      writer = csv.writer(file)
      writer.writerow(header)
  emotions = emotions.split()
  for g in emotions:
      for filename in os.listdir(f'/content/drive/MyDrive/Tesi/YoutubeAudioLibrary/Tracks/{g}'):
          songname = f'/content/drive/MyDrive/Tesi/YoutubeAudioLibrary/Tracks/{g}/{filename}'
            y, sr = librosa.load(songname, mono=True)

            y_harmonic, y_percussive = librosa.effects.hpss(y)

            tempo, beat_frames = librosa.beat.beat_track(y=y_harmonic, sr=sr)

            rmse = librosa.feature.rms(y=y)

            chroma=librosa.feature.chroma_cens(y=y_harmonic, sr=sr)

            cent = librosa.feature.spectral_centroid(y=y, sr=sr)
            spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)

            contrast=librosa.feature.spectral_contrast(y=y_harmonic,sr=sr)

            rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
            zcr = librosa.feature.zero_crossing_rate(y_harmonic)

            mfccs = librosa.feature.mfcc(y=y_harmonic, sr=sr, n_mfcc=13)

            chroma_mean=np.mean(chroma,axis=1)
            chroma_std=np.std(chroma,axis=1)

            #Generate the chroma Dataframe
            chroma_df=pd.DataFrame()
            for i in range(0,12):
                chroma_df['chroma_mean_'+str(i)]=chroma_mean[i]
            for i in range(0,12):
                chroma_df['chroma_std_'+str(i)]=chroma_mean[i]
            chroma_df.loc[0]=np.concatenate((chroma_mean,chroma_std),axis=0)


            mfccs_mean=np.mean(mfccs,axis=1)
            mfccs_std=np.std(mfccs,axis=1)

            #Generate the chroma Dataframe
            mfccs_df=pd.DataFrame()
            for i in range(0,13):
                mfccs_df['mfccs_mean_'+str(i)]=mfccs_mean[i]
            for i in range(0,13):
                mfccs_df['mfccs_std_'+str(i)]=mfccs_mean[i]
            mfccs_df.loc[0]=np.concatenate((mfccs_mean,mfccs_std),axis=0)

            cent_mean=np.mean(cent)
            cent_std=np.std(cent)
            cent_skew=scipy.stats.skew(cent,axis=1)[0]

            contrast_mean=np.mean(contrast,axis=1)
            contrast_std=np.std(contrast,axis=1)

            #Generate the chroma Dataframe
            contrast_df=pd.DataFrame()

            rolloff_mean=np.mean(rolloff)
            rolloff_std=np.std(rolloff)
            rolloff_skew=scipy.stats.skew(rolloff,axis=1)[0]

            spectral_df=pd.DataFrame()
            collist=['cent_mean','cent_std','cent_skew']
            for i in range(0,7):
                collist.append('contrast_mean_'+str(i))
            for i in range(0,7):
                collist.append('contrast_std_'+str(i))
            collist=collist+['rolloff_mean','rolloff_std','rolloff_skew']
            for c in collist:
                spectral_df[c]=0
            data=np.concatenate(([cent_mean,cent_std,cent_skew],contrast_mean,contrast_std,[rolloff_mean,rolloff_std,rolloff_skew]),axis=0)
            spectral_df.loc[0]=data

            zcr_mean=np.mean(zcr)
            zcr_std=np.std(zcr)
            zcr_skew=scipy.stats.skew(zcr,axis=1)[0]

            zrate_df=pd.DataFrame()
            zrate_df['zrate_mean']=0
            zrate_df['zrate_std']=0
            zrate_df['zrate_skew']=0
            zrate_df.loc[0]=[zcr_mean,zcr_std,zcr_skew]

            beat_df=pd.DataFrame()
            beat_df['tempo']=tempo
            beat_df.loc[0]=tempo

    
            to_append = pd.concat((chroma_df,mfccs_df,spectral_df,zrate_df,beat_df),axis=1).to_string(header=False,
                                                                                                      index=False,
                                                                                                      index_names=False)  
            
            to_append += f' {g}'

            file = open(savePath, 'a', newline='')
            with file:
                writer = csv.writer(file)
                writer.writerow(to_append.split())
```

# Script provato per l'analisi delle features utilizzate

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import f_classif


data = pd.read_csv('/content/drive/MyDrive/Tesi/YoutubeAudioLibrary/dataset2.csv')
data.head()# Dropping unneccesary columns
#data = data.drop(['filename'],axis=1)#Encoding the Labels
emotion_list = data.iloc[:, -1]
encoder = LabelEncoder()
y = encoder.fit_transform(emotion_list)#Scaling the Feature columns
scaler = StandardScaler()
X = data.iloc[:,0:73]

#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=f_classif, k=50)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(50,'Score'))  #print 10 best features
```

```python
import librosa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
from PIL import Image
import pathlib
import csv
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
import keras
from keras import layers
from keras.models import Sequential
import warnings
import scipy

from numpy.linalg import inv
warnings.filterwarnings('ignore')

def feature_extraction_windowed(savePath, emotions = 'arrabbiato calmo cupo drammatico felice funky motivante romantico triste vivace'):

  header = ""

  for i in range(0,12):
      header += 'chroma_mean_'+str(i)+' '
  for i in range(0,12):
      header += 'chroma_std_'+str(i)+' '

  for i in range(0,13):
      header += 'mfccs_mean_'+str(i)+' '
  for i in range(0,13):
      header += 'mfccs_std_'+str(i)+' '

  header += 'cent_mean cent_std cent_skew '

  for i in range(0,7):
      header += 'contrast_mean_'+str(i)+' '
  for i in range(0,7):
      header += 'contrast_std_'+str(i)+' '

  header += 'rolloff_mean rolloff_std rolloff_skew zrate_mean zrate_std zrate_skew tempo emotion'

  header = header.split()
  file = open(savePath, 'w', newline='')
  with file:
      writer = csv.writer(file)
      writer.writerow(header)
  emotions = emotions.split()
  count = 0
  for g in emotions:
      for filename in os.listdir(f'/content/drive/MyDrive/Tesi/YoutubeAudioLibrary/Tracks/{g}'):
          songname = f'/content/drive/MyDrive/Tesi/YoutubeAudioLibrary/Tracks/{g}/{filename}'
          track, sr = librosa.load(songname, mono=True)
          
          #normalized_track = librosa.util.normalize(track)
          count += 1
          print(str(count) + ")Nuova traccia: " + songname)
          #print("Dimensione traccia: " + str(track.shape))
          frame_len, hop_len = 337500, 225000 #15s window con 10s hop
          frames = librosa.util.frame(track, frame_length=frame_len, hop_length=hop_len, axis = 0)
        
          #print(frames.shape)
            
          for y in frames:
            
            y_harmonic, y_percussive = librosa.effects.hpss(y)

            tempo, beat_frames = librosa.beat.beat_track(y=y_harmonic, sr=sr)
            tempo = int(tempo)
            rmse = librosa.feature.rms(y=y)
            #rmse = librosa.util.normalize(rmse)

            chroma = librosa.feature.chroma_cens(y=y_harmonic, sr=sr)
            #chroma = librosa.util.normalize(chroma)

            cent = librosa.feature.spectral_centroid(y=y, sr=sr)
            #cent = librosa.util.normalize(cent)
            spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
            #spec_bw = librosa.util.normalize(spec_bw)

            contrast = librosa.feature.spectral_contrast(y=y_harmonic,sr=sr)
            #contrast = librosa.util.normalize(contrast)

            rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
            #rolloff = librosa.util.normalize(rolloff)
            zcr = librosa.feature.zero_crossing_rate(y_harmonic)
           # zcr = librosa.util.normalize(zcr)

            mfccs = librosa.feature.mfcc(y=y_harmonic, sr=sr, n_mfcc=13)
           # mfccs = librosa.util.normalize(mfccs)

            chroma_mean=np.mean(chroma,axis=1)
            chroma_std=np.std(chroma,axis=1)

            #Generate the chroma Dataframe
            chroma_df=pd.DataFrame()
            for i in range(0,12):
                chroma_df['chroma_mean_'+str(i)]=chroma_mean[i]
            for i in range(0,12):
                chroma_df['chroma_std_'+str(i)]=chroma_mean[i]
            chroma_df.loc[0]=np.concatenate((chroma_mean,chroma_std),axis=0)


            mfccs_mean=np.mean(mfccs,axis=1)
            mfccs_std=np.std(mfccs,axis=1)
            #Generate the chroma Dataframe
            mfccs_df=pd.DataFrame()
            for i in range(0,13):
                mfccs_df['mfccs_mean_'+str(i)]=mfccs_mean[i]
            for i in range(0,13):
                mfccs_df['mfccs_std_'+str(i)]=mfccs_mean[i]
            mfccs_df.loc[0]=np.concatenate((mfccs_mean,mfccs_std),axis=0)

            cent_mean=np.mean(cent)
            cent_std=np.std(cent)
            cent_skew=scipy.stats.skew(cent,axis=1)[0]

            contrast_mean=np.mean(contrast,axis=1)
            contrast_std=np.std(contrast,axis=1)

            #Generate the chroma Dataframe
            contrast_df=pd.DataFrame()

            rolloff_mean=np.mean(rolloff)
            rolloff_std=np.std(rolloff)
            rolloff_skew=scipy.stats.skew(rolloff,axis=1)[0]

            spectral_df=pd.DataFrame()
            collist=['cent_mean','cent_std','cent_skew']
            for i in range(0,7):
                collist.append('contrast_mean_'+str(i))
            for i in range(0,7):
                collist.append('contrast_std_'+str(i))
            collist=collist+['rolloff_mean','rolloff_std','rolloff_skew']
            for c in collist:
                spectral_df[c]=0
            data=np.concatenate(([cent_mean,cent_std,cent_skew],contrast_mean,contrast_std,[rolloff_mean,rolloff_std,rolloff_skew]),axis=0)
            spectral_df.loc[0]=data

            zcr_mean=np.mean(zcr)
            zcr_std=np.std(zcr)
            zcr_skew=scipy.stats.skew(zcr,axis=1)[0]

            zrate_df=pd.DataFrame()
            zrate_df['zrate_mean']=0
            zrate_df['zrate_std']=0
            zrate_df['zrate_skew']=0
            zrate_df.loc[0]=[zcr_mean,zcr_std,zcr_skew]

            beat_df=pd.DataFrame()
            beat_df['tempo']=tempo
            beat_df.loc[0]=tempo

    
            to_append = pd.concat((chroma_df,mfccs_df,spectral_df,zrate_df,beat_df),axis=1).to_string(header=False,
                                                                                                      index=False,
                                                                                                      index_names=False)  
            #print(to_append)
            to_append += f' {g}'

            file = open(savePath, 'a', newline='')
            with file:
                writer = csv.writer(file)
                writer.writerow(to_append.split())
```

```python
import librosa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
from PIL import Image
import pathlib
import csv
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
import keras
from keras import layers
from keras.models import Sequential
import warnings
import scipy

from numpy.linalg import inv
warnings.filterwarnings('ignore')

def metadata_extraction_windowed(savePath, emotions = 'arrabbiato calmo cupo drammatico felice motivante romantico triste vivace'):
  header = "file_name start end classID class"

  encoder = LabelEncoder()
  encoder = encoder.fit(emotions.split())
  file = open(savePath, 'w', newline='', encoding="utf-8")
  with file:
      writer = csv.writer(file)
      writer.writerow(header.split())
  emotions = emotions.split()
  count = 0
  for g in emotions:
      for filename in os.listdir(f'YoutubeAudioLibrary/Tracks/{g}'):
          songname = f'YoutubeAudioLibrary/Tracks/{g}/{filename}'
          track, sr = librosa.load(songname, mono=True)

          count += 1
          print(str(count) + ")Nuova traccia: " + songname)
          frame_len, hop_len = 337500, 225000 #15s window con 10s hop
          frames = librosa.util.frame(track, frame_length=frame_len, hop_length=hop_len, axis = 0)
        
          #print(frames.shape)
          start = 0
          end = int(frame_len/sr)
            
          for y in frames:
            
            
            to_append = [filename.encode("utf-8"), start, end, encoder.transform([g])[0], g]

            file = open(savePath, 'a', newline='', encoding="utf-8")
            with file:
                writer = csv.writer(file)
                writer.writerow(to_append)
            start += int(hop_len/sr)
            end = start + int(frame_len/sr)
```
